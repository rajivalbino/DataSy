### Database loja.db from sqlite2.ipynb

```python
import sqlite3
import matplotlib.pyplot as plt

conn = sqlite3.connect('files/loja.db')
c = conn.cursor()
```

```python
c.execute('SELECT ID, Value FROM Produtos')
ids = []
values = []
for row in c.fetchall():
    ids.append(row[0])
    values.append(row[1])

plt.plot(ids, values)
plt.xticks(ids)
plt.show()
```

```python
c.close()
conn.close()
```
